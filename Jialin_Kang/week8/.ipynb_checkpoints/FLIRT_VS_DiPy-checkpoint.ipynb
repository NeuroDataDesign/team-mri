```python
import nibabel as nib
import skimage.io as io
import numpy as np
from dipy.align.imaffine import (transform_centers_of_mass,
                                 AffineMap,
                                 MutualInformationMetric,
                                 AffineRegistration)
from dipy.align.transforms import (TranslationTransform3D,
                                   RigidTransform3D,
                                   AffineTransform3D)
from dipy.viz import regtools
```

```python
img_in = nib.load('e:/afterRun/t1w_brain_nores.nii.gz')
img_ref = nib.load('e:/MNI152_T1_2mm_brain.nii.gz')
omat = np.loadtxt('e:/afterRun/xfm_t1w2mni_init.mat')
img_in_init = nib.load('e:/t1w_brain_nores.nii.gz')

print(img_in.affine)
print(img_in.shape)
print('--'*30)
print(img_ref.affine)
print(img_ref.shape)
print('--'*30)
print(img_in_init.affine)
print(img_in_init.shape)
print('--'*30)
print(omat)
```

```python
# show the image in
img_in_arr = img_in.get_data()
io.imshow(img_in_arr[46])
```

```python
# show the image ref
img_ref_arr = img_ref.get_data()
io.imshow(img_ref_arr[46])
```

```python
# show the image in_init
img_in_init_arr = img_in_init.get_data()
io.imshow(img_in_init_arr[46])
```

```python
identity = np.eye(4)
print(img_in_init_arr.shape)
print(img_ref.shape)
print(img_in_init.affine)
print('--'*30)
affine_map = AffineMap(identity,
                       img_ref_arr.shape, img_ref.affine,
                       img_in_init_arr.shape, img_in_init.affine)
sam_1 = affine_map.transform(img_in_init_arr)
print(sam_1.shape)
print(img_ref.shape)
regtools.overlay_slices(img_ref_arr, sam_1, None, 0,
                        "com", "sam_1")
```

```python
c_of_mass = transform_centers_of_mass(img_ref_arr, img_ref.affine,
                                      img_in_init_arr, img_in_init.affine)
sam_2 = c_of_mass.transform(img_in_init_arr)
regtools.overlay_slices(img_ref_arr, sam_2, None, 0,
                        "com", "sam_2")
```

```python
nbins = 32
sampling_prop = None
metric = MutualInformationMetric(nbins, sampling_prop)

level_iters = [10000, 1000, 100]
sigmas = [3.0, 1.0, 0.0]
factors = [4, 2, 1]
affreg = AffineRegistration(metric=metric,
                            level_iters=level_iters,
                            sigmas=sigmas,
                            factors=factors)
transform = TranslationTransform3D()
params0 = None
starting_affine = c_of_mass.affine
print(starting_affine)
translation = affreg.optimize(img_ref_arr, img_in_init_arr, transform, params0,
                              img_ref.affine, img_in_init.affine,
                              starting_affine=starting_affine)
```

```python
transformed = translation.transform(img_in_init_arr)
regtools.overlay_slices(img_ref_arr, transformed, None, 0,
                        "ref", "Transformed")
print(translation.affine)
```

```python
identity = np.eye(4)
print(img_in_arr.shape)
print(img_ref.shape)
print(img_in.affine)
print('--'*30)
affine_map = AffineMap(identity,
                       img_ref_arr.shape, img_ref.affine,
                       img_in_arr.shape, img_in.affine)
sam_1 = affine_map.transform(img_in_arr)
print(sam_1.shape)
print(img_ref.shape)
regtools.overlay_slices(img_ref_arr, sam_1, None, 0,
                        "com", "sam_1")
```
